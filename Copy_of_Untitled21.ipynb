## Convert Data Types

### Subtask:
Convert relevant columns to their appropriate data types. Specifically, convert 'submitted_valuation', 'final_valuation', 'new_units', 'total_sqft', and 'stories' to numeric types, and 'set_up', 'under_review', 'issued', 'final' to datetime objects for time-based analysis.


```python
numeric_cols = ['submitted_valuation', 'final_valuation', 'new_units', 'total_sqft', 'stories']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

datetime_cols = ['set_up', 'under_review', 'issued', 'final']
for col in datetime_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print("Converted numeric columns to numeric type and datetime columns to datetime objects.")
print(df[numeric_cols + datetime_cols].info())
```

## Analyze Permit Status Distribution

### Subtask:
Analyze the distribution of permit 'status' to understand the common states of permits. Generate a bar plot to visualize the counts of each status, ensuring clear labels and a title.

## Analyze Numerical Distributions

### Subtask:
Analyze and visualize the distributions of 'submitted_valuation', 'final_valuation', 'new_units', and 'total_sqft' using appropriate plots (histograms, box plots) to understand their spread, central tendency, and identify outliers.

```python
numeric_cols_to_plot = ['submitted_valuation', 'final_valuation', 'new_units', 'total_sqft']

for col in numeric_cols_to_plot:
    plt.figure(figsize=(14, 6))

    # Histogram
    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    sns.histplot(df[col].dropna(), kde=True, bins=50, color='skyblue')
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('Frequency')

    # Box Plot
    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    sns.boxplot(y=df[col].dropna(), color='lightcoral')
    plt.title(f'Box Plot of {col.replace("_", " ").title()}')
    plt.ylabel(col.replace("_", " ").title())

    plt.tight_layout()
    plt.show()
```

## Analyze Permits by County

### Subtask:
Determine the distribution of permits across different 'county' values. Generate a bar plot to visualize this, with clear labels and a title to show permit activity by county.


```python
county_counts = df['county'].value_counts()

# Filter out any non-county string that might have been introduced, e.g., 'county' itself if it somehow ended up as a value
county_counts = county_counts[county_counts.index != 'county']

plt.figure(figsize=(10, 6))
sns.barplot(x=county_counts.index, y=county_counts.values, palette='coolwarm')
plt.xlabel('County')
plt.ylabel('Number of Permits')
plt.title('Distribution of Permits by County')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Analyze Permits Over Time

### Subtask:
Analyze the trend of permits over time using the 'set_up' date column. Visualize the number of permits set up per year or month using a line plot to identify any temporal patterns or seasonalities.


```python
permits_per_year = df['set_up'].dt.year.value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=permits_per_year.index, y=permits_per_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Permits')
plt.title('Permits Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Data preprocessing


```python
numerical_cols_for_outliers = ['submitted_valuation', 'final_valuation', 'new_units', 'total_sqft']

for col in numerical_cols_for_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    print(f"Outliers in '{col}' capped using IQR method. Upper bound: {upper_bound:.2f}")

print("\nOutlier treatment complete for specified numerical columns.")
```

## Feature Engineering: Time-Based Features

### Subtask:
Engineer new time-based features from the 'set_up' column to capture temporal patterns. Create columns for 'set_up_year', 'set_up_month', 'set_up_day_of_week', and 'set_up_quarter'.

```python
df['set_up_year'] = df['set_up'].dt.year
df['set_up_month'] = df['set_up'].dt.month
df['set_up_day_of_week'] = df['set_up'].dt.dayofweek
df['set_up_quarter'] = df['set_up'].dt.quarter

print("Time-based features engineered successfully.")
print(df[['set_up', 'set_up_year', 'set_up_month', 'set_up_day_of_week', 'set_up_quarter']].head())
```

```python
import pandas as pd

# Identify the categorical columns to encode
categorical_cols_to_encode = [
    'status',
    'type',
    'work',
    'neighborhood',
    'neighborhood_coalition',
    'business_association'
]

# Filter out columns that might not exist in the DataFrame
existing_categorical_cols = [col for col in categorical_cols_to_encode if col in df.columns]

# Apply One-Hot Encoding using pd.get_dummies()
# Use 'prefix' to make new column names identifiable
df_encoded = pd.get_dummies(df[existing_categorical_cols], prefix=existing_categorical_cols)

# Concatenate the one-hot encoded DataFrame with the original DataFrame
# and drop the original categorical columns
df = pd.concat([df, df_encoded], axis=1)
df = df.drop(columns=existing_categorical_cols)

print("Categorical features have been one-hot encoded.")

# Verify the transformation by inspecting the head and new column names
print("\nHead of the updated DataFrame:")
print(df.head())

print("\nNew columns added for categorical features (first 10 new columns):")
# Get all columns, then filter to find those starting with the prefixes
newly_encoded_cols = [col for col in df.columns if any(col.startswith(prefix) for prefix in existing_categorical_cols)]
print(newly_encoded_cols[:10])
```
